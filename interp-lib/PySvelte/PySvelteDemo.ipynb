## PySvelte demonstration

```python
import pysvelte
import numpy as np
```

```python
pysvelte.Hello(name='World')
```

```python
tokens = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']
attention_pattern = np.tril(np.random.normal(loc=0.3, scale=0.2, size=(8,8,3)))
pysvelte.AttentionMulti(tokens=tokens, attention=attention_pattern)
```

```python
tokens = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']
activations = np.random.normal(loc=0.3, scale=0.2, size=(8,))
pysvelte.TextSingle(tokens=tokens, activations=activations, neuron_name="Bob")
```
